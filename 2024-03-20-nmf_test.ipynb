> It's important to note that the solutions are stochastic and will vary with each run. The goal is explainable so all robust qualitative features remain regardless of non-unique solutions. 


```python
import numpy as np
from sklearn.decomposition import NMF
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(100, 50)
print("Shape of Base  matrix X: ", X.shape)
print("Matrix X:\n", X)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 2)
plt.imshow(X, cmap='magma', aspect='auto', vmin=0, vmax=1)
plt.title('Base Matrix X')
plt.colorbar()

plt.tight_layout()
plt.show()
```

```python
# Specify the number of components for factorization
n_components = 10

# Perform NMF
model = NMF(n_components=n_components)
W = model.fit_transform(X)
H = model.components_

print("Shape of Factorized matrix W: ", W.shape)
print("Shape of Factorized matrix H: ", H.shape)
```

```python
# Visualize the factorized matrices as heatmaps
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(W, cmap='magma', aspect='auto')
plt.title('Factorized Matrix W')
plt.colorbar()

plt.subplot(1, 2, 2)
plt.imshow(H, cmap='magma', aspect='auto')
plt.title('Factorized Matrix H')
plt.colorbar()

plt.tight_layout()
plt.show()
```

```python
X_2 = np.matmul(W,H)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 2)
plt.imshow(X_2, cmap='magma', aspect='auto', vmin=0, vmax=1)
plt.title('Reconstructed Matrix X_2')
plt.colorbar()

plt.tight_layout()
plt.show()
```
